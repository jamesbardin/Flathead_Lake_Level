##2023 Estimated Contribution to Surface Height

```
import requests
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

def fetch_usgs_streamflow_data(site_number, parameter_code="00065"):
    # Define the base URL for the USGS Instantaneous Values service
    base_url = "https://waterservices.usgs.gov/nwis/iv/"

    end_date = datetime.now()
    start_date = end_date - timedelta(days=120)

    # Parameters for the API request
    params = {
        "format": "json",
        "sites": site_number,
        "parameterCd": parameter_code,
        "startDT": start_date.strftime('%Y-%m-%d'),
        "endDT": end_date.strftime('%Y-%m-%d')
    }

    # Make the API request
    response = requests.get(base_url, params=params)
    response.raise_for_status()

    return response.json()

def extract_data_for_plotting(data):
    times = [entry['dateTime'] for entry in data['value']['timeSeries'][0]['values'][0]['value']]
    values = [float(entry['value']) for entry in data['value']['timeSeries'][0]['values'][0]['value']]

    times = [datetime.fromisoformat(time) for time in times]

    return times, values

def compute_cumulative_volume(values, times):
    volumes = []
    total_volume = 0

    # Compute the time difference in seconds between data points
    dt = (times[1] - times[0]).total_seconds()

    for value in values:
        total_volume += value * dt
        volumes.append(total_volume)

    return volumes

def compute_height_change(volumes, surface_area):
    return [volume / surface_area for volume in volumes]

def plot_data(times_list, values_list, labels):
    fig, ax1 = plt.subplots(figsize=(12,6))

    # Calculate cumulative volume for Kalispell and Polson sites
    kalispell_volume = compute_cumulative_volume(values_list[0], times_list[0])
    polson_volume = compute_cumulative_volume(values_list[1], times_list[1])

    # Calculate differential volume
    differential_volume = [v1 - v2 for v1, v2 in zip(kalispell_volume, polson_volume)]

    # Calculate height change due to the differential volume
    surface_area = 5.4920448e9  # surface area in square feet
    height_changes = compute_height_change(differential_volume, surface_area)

    ax1.plot(times_list[0], height_changes, '-o', markersize=0.5, label="Estimate Height Change (Kalispell - Polson) (based on 197mi^2 SA)")
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Height Change (ft)')
    ax1.legend(loc="upper left")

    # Plotting data for the third site on the right-hand side axis
    ax2 = ax1.twinx()
    times, values = times_list[2], values_list[2]
    ax2.plot(times, values, '-o', markersize=0.5, color="red", label=custom_labels[2])
    ax2.set_ylabel('Elevation', color="black")
    ax2.tick_params(axis='y', labelcolor="black")
    ax2.legend(loc="upper right")

    ax1.set_ylim(0, 15)
    ax2.set_ylim(2884.6, 2899.6)

    plt.title('Height Change and Elevation over the Last 120 Days')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()


custom_labels = ["Kalispell Gage", "Polson Gage", "Reservoir Elevation"]
site_numbers = ["12363500", "12372000", "12371550"]
parameter_codes = ["00060", "00060", "00062"]
times_list = []
values_list = []

for site, param_code in zip(site_numbers, parameter_codes):
    data = fetch_usgs_streamflow_data(site, param_code)
    times, values = extract_data_for_plotting(data)
    times_list.append(times)
    values_list.append(values)

plot_data(times_list, values_list, site_numbers)


```
